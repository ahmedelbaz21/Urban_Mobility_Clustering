### Task 1: Determine Business Objectives

The first objective of the data analyst is to thoroughly understand, from a business perspective, what the customer really wants to accomplish. This involves:

- **Background:** Identify the organisational context and the business area the project relates to.
- **Business Objectives:** Clearly state the primary question or problem the business wants to solve.
- **Business Success Criteria:** Define the measurable criteria that determine whether the project's outcome is considered a success from the business's point of view (e.g., "reduce customer churn by 15%", "improve prediction accuracy to 90%").

**Instructions:** In the cells below, define your business objectives, background context, and success criteria as Python variables (strings or dictionaries). This ensures your business logic is documented directly in the notebook's code state.

```python
business_objectives = {
    "background": "The project is set within the urban transportation and taxi services industry, specifically focusing on New York City's Yellow Taxi system. The dataset contains detailed records of taxi trips, including pickup times, locations, and trip characteristics. In a highly dynamic urban environment, demand for taxis varies significantly across different areas and times, making efficient driver allocation a key challenge for service providers.",

    "primary_objective": "The main objective is to identify high-demand pickup zones and temporal mobility patterns using clustering techniques on NYC Yellow Taxi trip data,\n in order to improve driver positioning, reduce passenger wait times, and enhance overall service efficiency.",

    "success_criteria": [
        "Criterion 1: Identify meaningful geographic clusters representing high-demand pickup zones",
        "Criterion 2: Uncover clear temporal patterns in taxi demand across hours of the day",
        "Criterion 3: Demonstrate differences in demand between weekdays and weekends",
        "Criterion 4: Generate actionable insights that could help optimize driver allocation and reduce idle time"
    ]
}
```

### Task 3: Determine Data Mining Goals

A business goal states objectives in business terminology. A data mining goal states project objectives in technical terms. This task converts the business objectives identified in Task 1 into specific, measurable data mining goals:

- **Data Mining Problem Type:** Classify the problem (e.g., classification, regression, clustering, anomaly detection, association rule mining, etc.).
- **Data Mining Goals:** State the technical objectives clearly (e.g., "Build a classification model to predict [target variable] with an F1-score ≥ 0.85").
- **Data Mining Success Criteria:** Define the technical metrics that will be used to evaluate the model's performance (e.g., accuracy, precision, recall, RMSE, silhouette score).

**Instructions:** Map your business objectives to concrete data mining goals.

```python
data_mining_goals = {
    "problem_type": ["Clustering"],

    "target_variable": None,

    "technical_goals": [
        "Apply clustering techniques (K-Means, DBSCAN) to identify high-demand and low-demand geographic zones based on pickup density.",
        "Analyze spatial patterns of Uber pickups using latitude and longitude.",
        "Analyze temporal patterns (hour, weekday) to understand demand fluctuations.",
        "Support demand forecasting and resource optimization through identified patterns."
    ],

    "success_metrics": [
        
            "Silhouette Score ≥ 0.5",
            "Davies-Bouldin Index minimized",
            "Clusters are interpretable and meaningful (high vs low demand zones)"
        
    ]
}
```

```python
import pandas as pd

rows = []

for key, value in data_mining_goals.items():
    if isinstance(value, list):
        content = "\n".join(f"• {item}" for item in value)

    elif isinstance(value, dict):
        content = "\n".join(
            f"• {k.replace('_', ' ').title()}: "
            + (", ".join(v) if isinstance(v, list) else str(v))
            for k, v in value.items()
        )

    else:
        content = value

    rows.append({
        "Section": key.replace("_", " ").title(),
        "Details": content
    })


rows.append({
    "Section": "Mapping",
    "Details": (
        "• Business objective: Improve ride efficiency → Regression (predict demand/trip duration)\n"
        "• Business objective: Reduce congestion → Clustering (identify geographic hotspots)\n"
        "• Business objective: Optimize driver allocation → Classification (low/medium/high demand zones)"
    )
})

df = pd.DataFrame(rows)

df.style.set_properties(**{
    'text-align': 'left',
    'white-space': 'pre-wrap',
    'vertical-align': 'top',
    'padding': '10px 16px'
}).hide(axis='index')
```
